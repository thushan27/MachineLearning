```python
"""
Agglomerative Clustering (bottom up hierarchical clustering) using Mall Customers
"""

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
customers = pd.read_csv("Datasets/Mall_Customers.csv")

customers.head()
```

```python
# Take annual income and spending score to find 5 clusters

x = customers[['Annual Income (k$)', 'Spending Score (1-100)']]

model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
model.fit(x)
```

```python
labels = model.labels_
customers['Labels'] = labels
```

```python
plt.scatter(x.iloc[:,0], x.iloc[:,1], c = customers['Labels'])

plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
```
